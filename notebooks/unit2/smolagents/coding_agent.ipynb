```python
from smolagents import CodeAgent, DuckDuckGoSearchTool, LiteLLMModel 
from dotenv import load_dotenv

load_dotenv('.env')

# Load Model from gemini
model = LiteLLMModel(model_id='gemini/gemini-2.0-flash')

# Define the Feature Selection Agent
agent = CodeAgent(
    tools=[DuckDuckGoSearchTool()],
    additional_authorized_imports=['bs4', 'BeautifulSoup', 'json'],
    model=model,
)

agent.run("""
    use beautifulsoup to extract all information from this page 
    'https://www.linkedin.com/in/abdallah-bakr/' then save data as data.json
""")
```
